```
import urllib.request
import json
import sqlite3


url = "https://data.nasa.gov/resource/gh4g-9sfh.json"
data = urllib.request.urlopen(url)

conn = sqlite3.connect('meteors.sqlite')
cur = conn.cursor()

conn.executescript('''
  DROP TABLE IF EXISTS Meteors;
  DROP TABLE IF EXISTS Fall;
  DROP TABLE IF EXISTS NameType;
  CREATE TABLE Meteors (name TEXT, id TEXT PRIMARY KEY, nametype INTEGER, mass INTEGER, fall INTEGER, year DATE, reclat TEXT, reclong TEXT, country INTEGER);
  CREATE TABLE Fall (id INTEGER PRIMARY KEY AUTOINCREMENT DEFAULT 1, fall TEXT);
  CREATE TABLE NameType (id INTEGER PRIMARY KEY AUTOINCREMENT DEFAULT 1, nametype TEXT);
''')

datajs = json.load(data)
#print(datajs)

for meteor in datajs:
    try:
        cur.execute("SELECT id FROM NameType WHERE nametype = (?)", (meteor['nametype'],))
        row = cur.fetchone()
        if row is None:
            cur.execute("INSERT INTO NameType (nametype) VALUES (?)", (meteor['nametype'],))
            type_id = cur.execute("SELECT id FROM NameType WHERE nametype = ?", (meteor['nametype'],))
        else:
            type_id = row[0]
    except Exception as e:
        print("error with type", meteor['id'], meteor['name'], e )
        continue

    try:
        cur.execute("SELECT id FROM Fall WHERE fall = (?)", (meteor['fall'],))
        row = cur.fetchone()
        if row is None:
            cur.execute("INSERT INTO Fall (fall) VALUES (?)", (meteor['fall'],))
            fall_id = cur.execute("SELECT id FROM Fall WHERE fall = (?)", (meteor['fall'],))
        else:
            fall_id = row[0]
    except Exception as e:
        print("error with fall", meteor['id'], meteor['name'], e )
        continue



    try:
        conn.execute("INSERT OR IGNORE INTO Meteors (name, id, nametype, mass, fall, year, reclat, reclong, country) values (?, ?, ?, ?, ?, ?, ?, ?, ?)", (meteor['name'], meteor['id'], type_id, meteor['mass'], fall_id, meteor['year'].split('T')[0], meteor['reclat'], meteor['reclong'], None))
        print("Data fed: ", meteor['id'],' ', meteor['name'], None)
    except:
        print("Unable to feed data: ", meteor['id'],' ', meteor['name'])

conn.commit()
```
